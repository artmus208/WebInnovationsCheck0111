```python
tokens = nltk.word_tokenize(data)

print(tokens)


name_tag_group_map = {
    #Существительное
    "Nouns": ['NN', 'NNP', 'NNPS', 'NNS'],
    #Прилагательное
    "Adjectives": ['JJ', 'JJR', 'JJS'],
    #Глаголы
    "Verbs": ['VB','VBP', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'],
    #Наречия
    "Adverbs": ["RBR", "RBS"],
    #Междометия
    "Interjections": ['IN'],
    #Предлоги
    "Prepositions": ["PRP", "PRPS"],
}

pos_lst = nltk.pos_tag(tokens)
print(pos_lst[:6])


# res = {}
# print(res)

for _el in pos_lst:
  # Если второй элемент из кортежа значений в pos_lst является частью речи,
  # которую мы учитываем, то увеличиваем соотв. результирующий счетчик res.
  # if _el[1] in

  pass

#   # или использовать Counter
#   # https://docs-python.ru/standart-library/modul-collections-python/klass-counter-modulja-collections/

invert_pos_lst = []

for _el in pos_lst:
  invert_pos_lst.append(  (_el[1], _el[0]) )

```

```python
invert_pos_lst[:10]
```

```python
count = 0

for el in pos_lst[:100]:
  # el: ('word', 'part_of_speech')
  if el[1] in name_tag_group_map['Nouns']:
    if el[1] in ["NNP", "NNPS"]:
      count+=1

print(count)
```

```python
from collections import Counter

simple_word_stats = Counter(invert_pos_lst)


simple_word_stats.most_common(50)

# print(simple_word_stats)
```
